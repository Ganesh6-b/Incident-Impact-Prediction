```python
def modelfit(alg, dtrain, predictors, performCV=True, printFeatureImportance=True, cv_folds=5):
    alg.fit(dtrain[predictors], dtrain['impact'])
        
    dtrain_predictions = alg.predict(dtrain[predictors])
    
    if performCV:
        cv_score = cross_val_score(alg, dtrain[predictors], dtrain['impact'], cv=cv_folds, scoring='roc_auc_ovr')
    
    print ("\nModel Report")
    print ("Accuracy : %.4g" % metrics.accuracy_score(dtrain['impact'].values, dtrain_predictions))
    
    if performCV:
        print ("CV Score : Mean - %.7g | Std - %.7g | Min - %.7g | Max - %.7g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
        
    if printFeatureImportance:
        feat_imp = pd.Series(alg.feature_importances_, predictors).sort_values(ascending=False)
        feat_imp.plot(kind='bar', title='Feature Importances')
        plt.ylabel('Feature Importance Score')


ser4 = ser2.drop(['impact'],axis=1)
predictors1 = [x for x in ser4.columns]
rafc1 = RandomForestClassifier(random_state=42)
modelfit(rafc1, ser2, predictors1)

```

```python
#################   2 Random forest classifier   ################# 
X2 = ser2.drop(['impact'],axis=1)
Y2 = ser2['impact']
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, Y2, test_size=0.33, random_state=42)
clf=RandomForestClassifier(class_weight='balanced')
rf_train = clf.fit(X2_train,Y2_train)
Y_pred2=rf_train.predict(X2_test)
score2=metrics.balanced_accuracy_score(Y2_test, Y_pred2)
print(score2)

print("F1_score :",f1_score(Y_pred2,Y2_test,average='weighted'))               
print("Precision_score :",precision_score(Y2_test, Y_pred2,average='weighted'))       
print("Recall:",recall_score(Y2_test,Y_pred2,average='weighted'))
print("classification_report", classification_report(Y2_test,Y_pred2))
confusion_matrix(Y2_test,Y_pred2)
```

```python

from sklearn.metrics import roc_curve, auc
from sklearn.svm import LinearSVC
from sklearn.preprocessing import label_binarize

Y2b = label_binarize(Y2, classes=[1,2,3])
Y2b_train, Y2b_test = train_test_split(Y2b, test_size=0.33, random_state=42)

clfb = RandomForestClassifier(class_weight='balanced')
y2b_score = clfb.fit(X2_train,Y2b_train)                 
Y_pred2b=y2b_score.predict(X2_test)
n_classes1 = 3

fpr1 = dict()
tpr1 = dict()
roc_auc = dict()
for i in range(n_classes1):
    fpr1[i], tpr1[i], _ = roc_curve(Y2b_test[:, i], Y_pred2b[:, i])
    roc_auc[i] = auc(fpr1[i], tpr1[i])

for i in range(n_classes1):
    plt.figure()
    plt.plot(fpr1[i], tpr1[i], label='ROC curve (area = %0.2f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate1')
    plt.ylabel('True Positive Rate1')
    plt.title('Receiver operating characteristic example1')
    plt.legend(loc="best")
    plt.show()
```
